```python
# ==============================
# SUBIR PROYECTO COMPLETO A GITHUB DESDE COLAB
# ==============================

import os
import subprocess
from getpass import getpass

# --- CONFIGURA ESTOS DATOS ---
PROJECT_PATH = "/content/drive/MyDrive/TelecomX_Churn_ML"
GITHUB_USER = "SantiAp11"
REPO_NAME = "TelecomX_Churn_ML"
GIT_EMAIL = "santiagoaparicioperez674@gmail.com"
GIT_NAME = "Santiago Aparicio"
COMMIT_MESSAGE = "Actualiza estructura completa del proyecto TelecomX Churn ML"
# -----------------------------

def run(cmd, check=True):
    print(f"\n$ {cmd}")
    result = subprocess.run(
        cmd,
        shell=True,
        text=True,
        capture_output=True
    )
    if result.stdout:
        print(result.stdout)
    if result.stderr:
        print(result.stderr)
    if check and result.returncode != 0:
        raise RuntimeError(f"Error ejecutando: {cmd}")
    return result

# 1) Ir al proyecto
os.chdir(PROJECT_PATH)
print("Directorio actual:", os.getcwd())

# 2) Inicializar git si hace falta
if not os.path.exists(os.path.join(PROJECT_PATH, ".git")):
    run("git init")

# 3) Configurar identidad git
run(f'git config user.name "{GIT_NAME}"')
run(f'git config user.email "{GIT_EMAIL}"')

# 4) Asegurar rama main
run("git branch -M main", check=False)

# 5) Configurar remoto limpio
remote_url = f"https://github.com/{GITHUB_USER}/{REPO_NAME}.git"
existing_remote = run("git remote", check=False).stdout.strip().split()

if "origin" in existing_remote:
    run(f"git remote set-url origin {remote_url}")
else:
    run(f"git remote add origin {remote_url}")

# 6) Añadir archivos
run("git add .")

# 7) Commit solo si hay cambios
status = run("git status --porcelain", check=False).stdout.strip()
if status:
    run(f'git commit -m "{COMMIT_MESSAGE}"', check=False)
else:
    print("\nNo hay cambios nuevos para commitear.")

# 8) Pedir token de forma oculta
token = getpass("Pega tu GitHub Personal Access Token: ").strip()

# 9) Usar remoto temporal con token
auth_remote = f"https://{GITHUB_USER}:{token}@github.com/{GITHUB_USER}/{REPO_NAME}.git"
run(f"git remote set-url origin {auth_remote}")

# 10) Push forzado
# Úsalo porque el repo remoto ya tenía un README inicial y quieres dejar tu versión completa
run("git push -u origin main --force")

# 11) Limpiar remoto para no dejar el token
run(f"git remote set-url origin {remote_url}")

print("\nProyecto subido correctamente a GitHub.")
print(f"Repositorio: {remote_url}")
```
