### Distribución de la variable objetivo

La variable `abandono` muestra que aproximadamente el 26.6% de los clientes cancelaron el servicio, mientras que el 73.4% permanecen activos. Esto indica un desbalance moderado entre clases.

Este comportamiento es consistente con el análisis exploratorio realizado previamente y deberá considerarse durante la evaluación de los modelos, priorizando métricas como recall y F1-score para la detección de clientes con riesgo de cancelación.

## Eliminación de columnas no predictivas

La columna `id_cliente` corresponde a un identificador único de cada cliente.  
Este tipo de variable no aporta valor al modelo predictivo y puede afectar el desempeño del algoritmo.

Por esta razón será eliminada del dataset antes del proceso de modelado.

```python
df = df.drop(columns=["id_cliente"])

df.head()
```
